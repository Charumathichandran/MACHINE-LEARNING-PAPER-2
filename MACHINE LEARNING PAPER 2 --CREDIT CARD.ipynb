### b. Outlier Treatment using the Z-score method


```python
import numpy as np
for i in df:
    mean = np.mean(df[i])
    std = np.std(df[i])
    df["zscore"]=(df[i]-mean)/std
    print('mean of the dataset is', mean)
    print('std of the dataset is', std)
    df['zscore'] = ( df[i] - df[i].mean() ) /df[i].std()
    df[df['zscore']>3]
    df[df['zscore']<-3]
    df_new = df[(df.zscore>-3) & (df.zscore<3)]
    print(df_new)


  
```

### Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5,random_state=42)
km.fit_predict(x)
```

```python
#WCSS=WITHIN CLUSTER SUM OF SQUARES
wcss=[]
for i in range(2,6):
    kmeans=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,tol=0.0001,verbose=0,random_state=None,copy_x=True,algorithm='auto')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
#ELBOW GRAPH
import matplotlib.pyplot as plt
plt.plot(range(2,6),wcss)
```

### Here from the ELBOW METHOD i found that optimal number of clusters=3

### Q5. Find the optimum value of k for k-means clustering using the silhouette score method and specify the number of observations in each cluster using a bar plot

```python
import sklearn.metrics as metrics
for i in range(3,10):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=200)
    kmeans.fit(df)
    labels=kmeans.labels_
    print("Silhouette  score for k(clusters)="+str(i)+" is "+str(metrics.silhouette_score(df,labels,metric="euclidean",sample_size=1000,random_state=200)))
```

### From the above observation i found that optimal number of clusters=3 by using Silhouette score

```python
#FROM THIS WE CAN OBSERVE THAT BOTH THE ELBOW METHOD AND SILHOUETTE SCORE GIVES SAME NUMBER OF OPTIMAL CLUSTERS AS 3.
```
